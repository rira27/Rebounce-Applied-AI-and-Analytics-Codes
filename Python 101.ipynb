**Problem 2 · The Currency Exchange**

```python
# Store INR amount and the exchange rate in two variables
inr = 75000
exchange_rate = 1.75

# Compute how many yen received
yen = inr * exchange_rate

# Print the result, formatted to two decimal places
print("Rs. 75000 converts to Yen", format(yen, ".2f"))
```
