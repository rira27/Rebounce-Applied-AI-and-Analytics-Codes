**Problem 1 · The Introduction Card**

```python
# Basic details
name = "Rishika Mehta"
city = "Udaipur"
year_of_birth = 1997

# Computing age
current_year = 2026
age = current_year - year_of_birth

print("Hi, I am", name, "from", city, ".This year I turn", age)
```

**Problem 2 · The Currency Exchange**

```python
# Store INR amount and the exchange rate in two variables
inr = 75000
exchange_rate = 1.75

# Compute how many yen received
yen = inr * exchange_rate

# Print the result, formatted to two decimal places
print("Rs. 75000 converts to Yen", format(yen, ".2f"))
```

**Problem 3 · The Grocery Bill**

```python
bread = 40
milk = 65.50
eggs = 120

# Compute the total value
total = bread + milk + eggs

# Compute the average value
average = total/3

# Print total and average value
print("Total = Rs.", format(total, ".2f"))
print("Average: Rs.", format(average, ".2f"))
```

**Problem 4 · Odd or Even?**

```python
# Store an integer
number = 88

# Check if the number is even
is_even = (number % 2 == 0)

# Print the number and the boolean together
print("Is 88 even?", is_even)
```
